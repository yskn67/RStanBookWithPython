```python
import pickle
import pystan
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("input/data-attendance-1.txt")
df.head()
```

```python
data = {
    "N": len(df),
    "A": df["A"],
    "Score": df["Score"] / 200,
    "Y": df["Y"],
}
stan_model = pystan.StanModel(file="model/model5-3.stan")
fit = stan_model.sampling(data=data, seed=1234)
```

```python
with open("output/result-model5-3.pkl", "wb") as f:
    pickle.dump(stan_model, f)
    pickle.dump(fit, f)
```
